```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sqlalchemy import create_engine
db= "SQL Challenge"
engine = create_engine(f"postgresql://postgres:grvrbw@localhost:5432/{db}")
connection = engine.connect()
```

```python
salarydata= pd.read_sql("SELECT * FROM salaries", connection)
print(salarydata)
```

```python
salary=salarydata.median
```

```python
salarydata.hist(column='salary')
plt.show()
```

```python
empdata= pd.read_sql("SELECT * FROM employees", connection)
print(empdata)
```

```python
titledata= pd.read_sql("SELECT * FROM titles", connection)
print(titledata)
```

```python
combo=pd.merge(empdata,salarydata, on="emp_no")
combo
```

```python
combotitle = combo.merge(titledata, left_on='emp_title_id', right_on='title_id')
combotitle
```

```python
# Create a bar chart of average salary by title
title=combotitle.groupby(by=["title"]).mean()
title
```

```python
title.plot.bar(y="salary", rot=45)
```
